```python
# Lista vacía
lista = []

while True:

    print("\n====== MENU LISTA ======")
    print("1. Agregar dato")
    print("2. Mostrar lista")
    print("3. Buscar dato")
    print("4. Eliminar dato")
    print("5. Contar datos")
    print("6. Limpiar lista")
    print("7. Salir")

    op = input("Selecciona una opcion: ")

    if op == "1":

        dato = input("Ingresa un dato: ")

        lista.append(dato)

        print("Dato agregado correctamente")

    elif op == "2":

        if lista == []:

            print("La lista esta vacia")

        else:

            print("\nDatos guardados:")

            i = 0

            while i < len(lista):

                print("-", lista[i])

                i += 1

    elif op == "3":

        buscar = input("Dato a buscar: ")

        if buscar in lista:

            print("El dato si existe")

        else:

            print("El dato no existe")

    elif op == "4":

        eliminar = input("Dato a eliminar: ")

        if eliminar in lista:

            lista.remove(eliminar)

            print("Dato eliminado")

        else:

            print("Ese dato no existe")

    elif op == "5":

        print("Cantidad de datos:", len(lista))

    elif op == "6":

        lista = []

        print("Lista limpiada")

    elif op == "7":

        print("Fin del programa")

        break

    else:

        print("Opcion incorrecta")
```
