```python
analistas = []

while True:

    op = input("1 agregar 2 mostrar 3 salir: ")

    if op == "1":
        analistas.append(input("Nombre: "))

    elif op == "2":
        print(analistas)

    elif op == "3":
        break
```

```python
analistas = []

while True:

    op = input("1 agregar 2 buscar 3 salir: ")

    if op == "1":
        analistas.append(input("Nombre: "))

    elif op == "2":

        if input("Buscar: ") in analistas:
            print("Si")

        else:
            print("No")

    elif op == "3":
        break
```

```python
analistas = []

while True:

    op = input("1 agregar 2 eliminar 3 salir: ")

    if op == "1":
        analistas.append(input("Nombre: "))

    elif op == "2":

        x = input("Eliminar: ")

        if x in analistas:
            analistas.remove(x)

    elif op == "3":
        break
```
